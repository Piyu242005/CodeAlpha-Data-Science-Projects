## ðŸ§¹ Step 3: Preparation & Engineering

We transition standard elements into algorithmic values:
- Generating `Car_Age` subtracting `Year` from the constant `2024`.
- Purging redundant tracking details.
- Converting categories (`Fuel_Type`, `Transmission`, etc.) through deep encoding sequences.

```python
# 1. Feature Genesis
df['Car_Age'] = 2024 - df['Year']
df['Depreciation'] = df['Present_Price'] - df['Selling_Price']

# 2. Integrity Cleaning
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)

df_model = df.copy()
df_model.drop(['Car_Name', 'Year'], axis=1, inplace=True)

# 3. Encoding Strings representing object constraints
categorical_cols = ['Fuel_Type', 'Selling_type', 'Transmission']
le = LabelEncoder()
enc_map = {}

for axis in categorical_cols:
    df_model[axis + '_Encoded'] = le.fit_transform(df_model[axis])
    df_model.drop(axis, axis=1, inplace=True)

display(df_model.head(3))
```

## ðŸ¤– Step 5: Advanced Model Generation

We test 6 independent algorithms, spanning Linear models down to heavy-duty decision trees predicting identical features to seek out the optimal regressor.

```python
TARGET = 'Selling_Price'
X = df_model.drop(['Selling_Price', 'Depreciation'], axis=1)
y = df_model[TARGET]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
print("Dimensions Ready for Injection.")
```

```python
models = {
    'Linear Matrix': LinearRegression(),
    'Ridge Matrix':  Ridge(alpha=1.0),
    'Lasso Matrix':  Lasso(alpha=0.1),
    'Decision Tree':     DecisionTreeRegressor(max_depth=6, random_state=101),
    'Random Forest':     RandomForestRegressor(n_estimators=100, random_state=101),
    'Gradient Boosting': GradientBoostingRegressor(n_estimators=100, random_state=101)
}

results = []
best_model_data = None

for name, model in models.items():
    model.fit(X_train, y_train)
    pred = model.predict(X_test)
    acc  = r2_score(y_test, pred)
    rms  = np.sqrt(mean_squared_error(y_test, pred))
    results.append({'Pipeline': name, 'RÂ² Index': acc, 'RMSE Float': rms})
    
    if name == 'Random Forest':
        best_model_data = model
        
pd.DataFrame(results).sort_values(by='RÂ² Index', ascending=False).style.background_gradient(cmap='Blues', subset=['RÂ² Index'])
```
